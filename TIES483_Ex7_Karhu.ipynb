### 7.1
Show

1) that an optimal solution to the weighting method is Pareto optimal, if the weights $w_i>0$ for all $i=1,\ldots,k$, and

2) that an optimal solution $x^*$ to the epsilon constraint method is Pareto optimal, if it is an optimal solution to the epsilon constraint method for all $j=1,\ldots,k$ and $\epsilon=f(x^*)$.


1)Weighting method is Pareto optimal


Weighting method is as follow:
$$
\min_{x\in S} \sum_{i=1}^kw_if_i(x)
$$ for different weights $w_i\geq0$, $i=1,\ldots,k$ such that $\sum_{i=1}^k w_i=1$.


Let's assume that a unique solution to weighting method is $x^*$ is not Pareto optimal, so there does exist a feasible solution $x$, such that 
$$
\left\{
\begin{align}
f_i(x)\leq f_i(x^*)\text{ for all }i\in \{1,\ldots,k\}\\
f_j(x)<f_j(x^*)\text{ for some }j\in \{1,\ldots,k\}.\\
\end{align}
\right.
$$

But as the assumptions the weigts are $w_i$ > 0, then for all x we have:
$$
\sum_{i=1}^kw_if_i(x) \leq \sum_{i=1}^kw_if_i(x^*)
$$

But because all weights were assumpted to be unique, the $x^*$ is unique, then for all $\hat{x} \in R$ we have:

$$
\sum_{i=1}^kw_if_i(x^*) < \sum_{i=1}^kw_if_i(\hat{x})
$$


And this is a conflict to the assumption and thus the optimal solution $x^*$ is Pareto Optimal.

2)Epsilon-constraint method is Pareto optimal:

Epsilon-constraint method is now:

$$
\begin{align}
\min \quad f_j(x)\\
\text{s.t. }\quad x\in S\\
f_i(x)\leq \epsilon_i \text{ for all }i\neq j
\end{align}
$$ for different bounds $\epsilon_i, i\neq j$.

Let's assume it's not Pareto Optimal, so there does exist a feasible solution $x$, for a epsilon-constraint solution  $x^*$ such that 
$$
\left\{
\begin{align}
f_i(x)\leq f_i(x^*)\text{ for all }i\in \{1,\ldots,k\}\\
f_j(x)<f_j(x^*)\text{ for some }j\in \{1,\ldots,k\}.\\
\end{align}
\right.
$$

Thus the solution to the epsilon-constraint problem for all $j=1,\ldots,k$ and $\epsilon < f(x^*)$.

But this brakes the epsilon-constraint, as it has to be $\epsilon \leq f(x^*)$. So the epsilon-constraint needs to be $\epsilon = f(x^*)$ and thus the problem in then Pareto Optimal.

### 7.3
Think of a multiobjective optimization problem from your own life with at least three objectives. Model this problem using algebraic equations.


Maximizing time usage of day:

In a day you only have 24 hours to do all the things you want to get done. This multiobjective optimization problem
is to maximize the activity during day-time such that studying, work and free time can be achieved all in one day, but you still want to be able to sleep properly at night and eat at day so that you are able to repeat this work habit everyday and aren't too tired or hungry after one day. Thus sleep, freetime and eating are minimized.

Constraints are the following:
$$
x1 => 6 \leq sleep \leq 10 \\
x2 => 1 \leq work \leq 10 \\
x3 => 1 \leq study\leq 6 \\
x4 =>  freetime \geq 0 \\
x5 => 1 \leq eat \leq 5
$$

So the minimize functions with constaints are the following
$$
min \{x1,-x2,-x3,x4,x5 \} \\
s.t. 6 \leq x1 \leq 10\\
1 \leq x2 \leq 10\\
1 \leq x5 \leq 5\\
1 \leq x3 \leq 6 \\
x4 \geq 0 \\
x1 + x2 + x3 + x4 + x5 \leq 24 $$
